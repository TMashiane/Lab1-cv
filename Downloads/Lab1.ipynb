````python
import cv2 as cv
import numpy as np
import scipy.stats
import skimage
import imageio
import mpmath
import matplotlib.pyplot as plt
import seaborn as sns
import PIL
````

````python
image1= cv.imread("/home/vmuser/Downloads/Puzzle Images Masks - Handout 1/image-35.jpg")[:,:,::-1]
image1.shape
plt.imshow(image1)
````

````python
#showing the 3 components of the image1
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(image1[:,:,0], cmap='Reds')
plt.title('Red Channel')

plt.subplot(1,3,2)
plt.imshow(image1[:,:,1], cmap='Greens')
plt.title('Green Channel')

plt.subplot(1,3,3)
plt.imshow(image1[:,:,2], cmap='Blues')
plt.title('Blue Channel')
plt.show()
````

````python
#converting the image1 to grayscale
gray_image= skimage.color.rgb2gray(image1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()

# Checking the shape of the grayscale image
gray_image.shape
````

````python
#converting the image1 to hsv color space
hsv_image = skimage.color.rgb2hsv(image1)
plt.imshow(hsv_image)
plt.title('HSV Image')
plt.show()

# Checking the shape of the hsv image
hsv_image.shape
````

````python
#Question1: What is the width and height of the image1?
width, height = image1.shape[1], image1.shape[0]
print(f"Width: {width}, Height: {height}")
````

````python
#Question2: How many white pixels are there in the image1?
white= np.array([255, 255, 255])
white_pixels = np.sum(np.all(image1 == white, axis=-1))
#white_pixels = np.sum(image1 == 255)
print(f"Number of white pixels: {white_pixels}")
````

````python
#Question 3: What is the maximum pixel value in the image1?
max_pixel_gray= np.max(gray_image)
print(f"Maximum pixel value in grayscale image: {max_pixel_gray}")
````

````python
#Question4: What is the maximum pixel value across all the puzzle pixels?
max_pixel_image1 = np.max(image1)
max_pixel_hsv = np.max(hsv_image)
answer = max(max_pixel_image1, max_pixel_hsv, max_pixel_gray)
print(f"Maximum pixel across all the puzzle pixels: {answer}")

````

````python
#Question5: What is mean(brightness) pixel value of the grayscale?
mean_brightness_gray = np.mean(gray_image)
print(f"Mean brightness value of the grayscale: {mean_brightness_gray}")
````

````python
#Question6: What is the mean brightness of the puzzle pixels?
mean_brightness = np.mean(image1)
print(f"Mean brightness of the puzzle pixels: {mean_brightness}")
````

````python
#Question7: What is the mean brightness of the background pixels?

back_image1= cv.imread("/home/vmuser/Downloads/Puzzle Images Masks - Handout 1/mask-35.png")[:,:,::-1]
plt.imshow(back_image1)

#Finding the mean brightness of the background pixels
mean_brightness_background = np.mean(back_image1)
print(f"Mean brightness of the background pixels: {mean_brightness_background}")
````

````python
#Q8. What is the variance in the grayscale intensities for puzzle pixels?
var_gray = np.var(gray_image)
print(f"Variance of the grayscale image: {var_gray}")
````

````python
#Q9. What is the variance in the grayscale intensities for background pixels?
var_background = np.var(back_image1)
print(f"Variance of the background pixels: {var_background}")
````
